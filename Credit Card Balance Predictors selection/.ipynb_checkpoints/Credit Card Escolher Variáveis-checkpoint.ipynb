# Regressão linear
## Seleção de preditores de Balance
#### Credit Card Balance Data


- O dataset de credit card balance contém informação dos clientes com cartão de crédito incluindo dados bancários e pessoais. <br>
- O objetivo desse notebook é utilizar a regressao linear para prever o Balance (Dívida) no cartão de crédito, selecionando os melhores preditores para um modelo _robusto e generalizável_.

```python
get_ipython().run_line_magic('matplotlib', 'inline')
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
import statsmodels.api as sm
sns.set_palette("twilight_shifted_r")
```

```python
data.isna().sum().sum()
```

```python
data.drop('Unnamed: 0', axis=1, inplace = True)
data.set_index('ID', inplace = True)
```

#### Transformando variáveis string em categóricas

```python
data['Student'] = data['Student'].replace({'Yes':1,'No':0})
data['Married'] = data['Married'].replace({'Yes':1,'No':0})
data['Ethnicity'] = data['Ethnicity'].replace({'Caucasian':'CA','Asian':'AS','African American':'AA'})
```

## Regressão Linear
### Escolha das variáveis
A inclusão de todas as variáveis no modelo explica praticamente toda a variância de Balance. Por isso, o modelo com certeza está com overfitting e algumas variáveis precisam sair para deixar o modelo mais generalizável

```python
model1 = sm.OLS.from_formula("Balance ~ Income + Limit + Rating + Cards + Education + Age + Ethnicity + Gender + Student + Married ", data=data)
result1 = model1.fit()
result1.summary()
```

1 - PVALUES
- Ethnicity e Gender possuem os maiores P valores e t statistics, podendo indicar que não são bons preditores para Balance
- Retirá-las do modelo não alterou o R² e aumentou a F Stat. 


```python
model1 = sm.OLS.from_formula("Balance ~ Income + Limit + Rating + Cards + Education + Age + Student + Married ", data=data)
result1 = model1.fit()
result1.summary()
```

Algumas variáveis mostraram sinais claros de multicolinearidade: Income, Limit e Rating.
Vamos excluir Rating por ser a variável menos 'pura' e resultado de avaliações internas do banco, não necessariamente do comportamento de consumo do cliente em si
- o r² manteve-se altissimo, não prejudicando a explicabilidade do nosso modelo

```python
model1 = sm.OLS.from_formula("Balance ~ Income + Limit + Cards + Education + Age + Student + Married ", data=data)
result1 = model1.fit()
result1.summary()
```

Education possui um p value altissimo e sua exclusão também não afetou o modelo
Além disso, as correlações de education com as demais variaveis eram baixas


```python
model1 = sm.OLS.from_formula("Balance ~ Income + Limit + Cards + Age + Student + Married ", data=data)
result1 = model1.fit()
result1.summary()
```

Age também mostra sinais de colinearidade, já que possui correlação com Income. Sua exclusão do modelo também não afetou grandemente a métrica de performance r²

```python
model1 = sm.OLS.from_formula("Balance ~ Income + Limit + Cards + Student + Married", data=data)
result1 = model1.fit()
result1.summary()
```

A próxima variável, Married, agora possui alto pvalor e por isso vamos descartá-la

```python
model1 = sm.OLS.from_formula("Balance ~ Income + Limit + Student + Cards", data=data)
result1 = model1.fit()
result1.summary()
```
